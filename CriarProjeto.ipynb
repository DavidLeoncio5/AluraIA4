```python
!pip install -q -U google-generativeai
import google.generativeai as genai

GOOGLE_API_KEY="AIzaSyC5U8OF19B0VlyU1g8lNDX4D4Ww0ZsQG0U"
genai.configure(api_key=GOOGLE_API_KEY)
import random

class Projeto:
  def __init__(self, titulo, criador, formato, tema, descricao):
    self.titulo = titulo
    self.criador = criador
    self.formato = formato
    self.tema = tema
    self.descricao = descricao
    self.curtidas = 0

  def apresentar(self):
    print("-" * 30)
    print(f"Título: {self.titulo}")
    print(f"Criador: {self.criador}")
    print(f"Formato: {self.formato}")
    print(f"Tema: {self.tema}")
    print(f"Descrição: {self.descricao}")
    print(f"Curtidas: {self.curtidas}")
    print("-" * 30)

  def curtir(self):
    self.curtidas += 1

formatos_disponiveis = ["Apresentação", "Vídeo", "Podcast", "Jogo", "Simulação", "Outro"]
temas_disponiveis = ["Ciência", "Tecnologia", "Engenharia", "Artes", "Matemática", "Outro"]

projetos = []

def criar_projeto():
  titulo = input("Título do projeto: ")
  criador = input("Nome do criador: ")

  print("\nFormatos disponíveis:")
  for i, formato in enumerate(formatos_disponiveis):
    print(f"{i+1}. {formato}")
  while True:
    opcao_formato = input("Escolha o formato (número): ")
    try:
      opcao_formato = int(opcao_formato)
      if 1 <= opcao_formato <= len(formatos_disponiveis):
        formato = formatos_disponiveis[opcao_formato - 1]
        break
      else:
        print("Opção inválida. Por favor, escolha um número da lista.")
    except ValueError:
      print("Entrada inválida. Por favor, digite um número.")

  print("\nTemas disponíveis:")
  for i, tema in enumerate(temas_disponiveis):
    print(f"{i+1}. {tema}")
  while True:
    opcao_tema = input("Escolha o tema (número): ")
    try:
      opcao_tema = int(opcao_tema)
      if 1 <= opcao_tema <= len(temas_disponiveis):
        tema = temas_disponiveis[opcao_tema - 1]
        break
      else:
        print("Opção inválida. Por favor, escolha um número da lista.")
    except ValueError:
      print("Entrada inválida. Por favor, digite um número.")

  descricao = input("Descrição do projeto: ")

  projeto = Projeto(titulo, criador, formato, tema, descricao)
  projetos.append(projeto)
  print("Projeto criado com sucesso!")

def explorar_projetos():
  if not projetos:
    print("Ainda não há projetos cadastrados.")
    return

  print("\nProjetos disponíveis:")
  for i, projeto in enumerate(projetos):
    print(f"{i+1}. {projeto.titulo} ({projeto.formato}, {projeto.tema})")

  while True:
    opcao_projeto = input("Escolha um projeto para visualizar (número) ou digite 'voltar' para retornar: ")
    if opcao_projeto.lower() == 'voltar':
      return
    try:
      opcao_projeto = int(opcao_projeto)
      if 1 <= opcao_projeto <= len(projetos):
        projeto_escolhido = projetos[opcao_projeto - 1]
        projeto_escolhido.apresentar()
        while True:
          acao = input("Digite 'curtir' para curtir o projeto, 'voltar' para retornar à lista ou 'sair' para finalizar: ").lower()
          if acao == 'curtir':
            projeto_escolhido.curtir()
            print("Você curtiu este projeto!")
          elif acao == 'voltar':
            break
          elif acao == 'sair':
            exit()
          else:
            print("Opção inválida.")
        break
      else:
        print("Opção inválida. Por favor, escolha um número da lista.")
    except ValueError:
      print("Entrada inválida. Por favor, digite um número.")


while True:
  print("\n### Plataforma de Projetos Inovadores ###")
  print("1. Criar novo projeto")
  print("2. Explorar projetos")
  print("3. Sair")

  opcao = input("Escolha uma opção (número): ")

  if opcao == '1':
    criar_projeto()
  elif opcao == '2':
    explorar_projetos()
  elif opcao == '3':
    print("Saindo da plataforma. Até mais!")
    break
  else:
    print("Opção inválida. Por favor, escolha um número da lista.")
```
